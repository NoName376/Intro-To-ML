# Q2. Given vector a = [5, -3] and scalar $\lambda$ =-2:

```python
a = np.array([5, -3])
lam = -2
a = lam * a
print(a)
```

```python
norm_a = np.linalg.norm(a)
norm_lam_a = np.linalg.norm(lam * a)
print(f"Norm a: {norm_a}, Norm lambda * a: {norm_lam_a}")
```

Multiplying by $\lambda $ doubled the length of the vector and changed its direction because it is negative

# Q9: Compute the Euclidean distance matrix between every point in group A and every point in group B.

```python
A = np.array([
 [ 2, 1],
 [ 5, 4],
 [ 1, 6],
 [ 7, 2]
])
```

```python
B = np.array([
 [0, 0],
 [6, 3],
 [3, 5]
])
```

```python
A_n = A[:, np.newaxis, :]
B_n = B[np.newaxis, :, :]
```

```python
diff = A_n - B_n
sq_diff = diff ** 2
```

```python
sum_sq = np.sum(sq_diff, axis=2)
dist_matrix = np.sqrt(sum_sq)
```

```python
print(dist_matrix)
```
