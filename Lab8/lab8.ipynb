## 4- What is the probability of a rainfall more than 500mm in June?


```python
total = df.shape[0]
```

```python
p_june_high = len(df[df['June_more_than_500'] == 1]) / total
p_june_high
```

## 5- What is the probability of having a flood?

```python
p_flood = len(df[df['FLOODS_BINARY'] == 1]) / total
p_flood
```

## 6- What is the probability of having a flood and a rainfall more than 500mm in June?


```python
p_flood_and_june = len(df[(df['FLOODS_BINARY'] == 1) & (df['June_more_than_500'] == 1)]) / total
p_flood_and_june
```

## 7- What is the Probability of having a flood given it rained more than 500mm in June?


```python
p_flood_given_june_high = p_flood_and_june / p_june_high
p_flood_given_june_high
```

## 8- What is the probability of a rainfall less than 500mm in June?

```python
p_june_low = 1 - p_june_high
p_june_low
```
