# Assignment 3: Understanding Customer Tipping Behavior

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Q1: Loading and Understanding the Dataset

```python
# Q1.1: Load dataset
df = sns.load_dataset('tips')
```

```python
# Q1.2: Display first rows
df.head()
```

```python
# Q1.3: Dataset shape and columns
print(df.shape)
print(df.columns.tolist())
```

### Q2: Basic Filtering - Who Tips More?

```python
# Q2.1: Weekend
weekend = df[df['day'].isin(['Sat', 'Sun'])]
```

```python
# weekday
weekday = df[df['day'].isin(['Sat', 'Sun'])]
```

```python
# Q2.2: Average tip, Average total bill
print("Weekend Avg Tip:", weekend['tip'].mean())
print("Weekend Avg Bill:", weekend['total_bill'].mean())

print("Weekday Avg Tip:", weekday['tip'].mean())
print("Weekday Avg Bill:", weekday['total_bill'].mean())
```

**Answer Q2.3: Compare this with weekday customers.**

I see that weekend customers pay more money. Their total bill is bigger (22.47 vs 18.28). Also, they give bigger tips (3.12 vs 2.78). Weekdays are cheaper.

### Q3: Sorting and Identifying High-Value Customers

```python
# Q3.1: Sort customers by total_bill (descending)
sorted_df = df.sort_values(by='total_bill', ascending=False)
```

```python
# Q3.2: Display the top 10 highest bills
sorted_df.head(10)
```

**Answer Q3.3: What do you notice about their group sizes?**

The groups are big. I see sizes 3, 4, and 6 in the top 10 list. If the group is small (size 2), the bill is usually smaller. So, big groups pay more.

### Q5: Visualization - What Influences Tips?

```python
# Q5.1: Tip vs Total Bill (scatter plot)
sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
# Q5.1: Average tip by day (bar plot)
sns.barplot(data=df, x='day', y='tip', errorbar=None)
plt.show()
```

```python
# Q5.1: Tip distribution (histogram)
sns.histplot(df['tip'], kde=True)
plt.show()
```

**Answer Q5.2: Interpret each plot in 1-2 sentences.**

* **Scatter:** If the bill is high, the tip is high. They go up together.
* **Bar:** Sunday is the best day. People give the biggest tips on Sunday and Saturday.
* **Histogram:** Most tips are small (2 to 4). Big tips are very rare.

### Q6: Final Insight & Conclusion

**Answer Q6:**

* **Who tips more?** People with big bills and big groups tip more.
* **When tips are highest?** Tips are highest on the weekend (Saturday and Sunday).
* **How this insight could help restaurant management:** Managers should put more workers on weekends. Also, waiters should try to serve big groups well because they pay more money.
