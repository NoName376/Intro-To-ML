```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'chipotle.tsv'
chipo = pd.read_csv(path, sep = '\t')
```

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

# 31. What was the most ordered item in the choice_description column?


```python
c1 = chipo.groupby('choice_description')[['quantity']].sum()
c1 = c1.sort_values(['quantity'], ascending=False)
c1.head(1)
```

# 32. What is the average revenue amount per order?

```python
c2 = chipo.groupby('order_id')['item_price']
c2.mean()
```

# 33. Which product has the highest total quantity sold?


```python
c3 = chipo.groupby('item_name')['quantity'].sum()
c3 = c3.sort_values(ascending=False)
c3.head(1)
```

# 34. calculate the average price of each product.

```python
chipo['one_item_price'] = chipo['item_price'] / chipo['quantity']

chipo.groupby('item_name')['one_item_price'].mean()
```

# 35. Which choice appears most often?

```python
c5 = chipo.groupby('choice_description').size()
c5 = c5.sort_values(ascending=False)
с5 = c5.head(1)
с5.index
```
